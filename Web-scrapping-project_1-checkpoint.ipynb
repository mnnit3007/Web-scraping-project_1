# To fetch top movies names,years and ratings to prepare a dataset by using this data

```python
# importing libraries
```

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
#Sending request to website to get html content of the website, which is stored in soup object
```

```python
url=requests.get('https://www.imdb.com/chart/top/?ref_=nv_mv_250',headers={'User-agent':'Mozilla/5.0'})
```

```python
soup=BeautifulSoup(url.text,'html.parser')
```

```python
#title object to store titles of movies using find_all() function.returns bs4.element.ResultSet
```

```python
Title=[] #List object to store the titles
```

```python
#Tags and class obtained from website html inspection.
title=soup.find_all('div',{"class":"ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-43986a27-9 gaoUku cli-title"})
```

```python
for i in range(len(title)):
    Title.append(title[i].find('h3').text) #find function to return tags, .text to get text part from html.
    print(title[i].find('h3').text) 
```

```python
#year object to store movies years of relears using find_all() function,returns bs4.element.ResultSet,which can be traversed by indexing
```

```python
year=soup.find_all('div',{"class":"sc-43986a27-7 dBkaPT cli-title-metadata"})
```

```python
Year=[]
```

```python
len(year)
```

```python
for i in range(len(year)):
    Year.append(year[i].find('span',{"class":"sc-43986a27-8 jHYIIK cli-title-metadata-item"}).text)
    print(year[i].find('span',{"class":"sc-43986a27-8 jHYIIK cli-title-metadata-item"}).text)
```

```python
#title object to store ratings of movies using find_all() function.returns bs4.element.ResultSet
```

```python
rating=soup.find_all('div',{"class":"sc-9ab53865-0 bIaPYM sc-43986a27-2 bvCMEK cli-ratings-container"})
```

```python
Rating=[]
```

```python
len(rating)
```

```python
for i in range(len(rating)):
    Rating.append(rating[i].find('span').text)
    print(rating[i].find('span').text)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Title':Title,'Year':Year,'Rating':Rating}) #df dataframe using the three lists
```

```python
df.to_csv("imdb_subdf.csv") #df saved into imdb_subdf csv file
```
